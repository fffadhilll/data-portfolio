## Melihat data null

### Fungsi check data null

```python
def missing_values_table(df):
    mis_val = df.isnull().sum()
    mis_val_percent = 100 * df.isnull().sum() / len(df)
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
    mis_val_table_ren_columns = mis_val_table.rename(
    columns = {0 : 'Missing Values', 1 : '% of Total Values'})
    mis_val_table_ren_columns = mis_val_table_ren_columns[mis_val_table_ren_columns.iloc[:,1] != 0].sort_values('% of Total Values', ascending=False).round(1)
    print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"
        "There are " + str(mis_val_table_ren_columns.shape[0]) +
            " columns that have missing values.")
    return mis_val_table_ren_columns
```

```python
sns.histplot(data=df2, x='total_duration_stays', binwidth=6)

plt.show()
```

```python
df2_grpby = df2.groupby(['hotel', 'is_canceled', 'total_duration_stays'])['lead_time'].count().reset_index(name='total_bookings')

df2_pivot = df2_grpby.pivot_table(index=['hotel', 'total_duration_stays'], columns='is_canceled', values='total_bookings').reset_index()

df2_pivot['pct_canceled'] = round(df2_pivot[1]*100.0/(df2_pivot[0]+df2_pivot[1]), 2)

df2_pivot
```

Karena total menginap mempunyai banyak kategori, maka Total menginap yang mencapai 3 minggu dan lebih dari 3 minggu akan dikelompokkan

- 15 = lebih dari 14 malam
- 21 = lebih dari 3 minggu

```python
df2['total_duration_stays_group'] = np.where(df2['total_duration_stays'] > 21, 21,
                                                   np.where(df2['total_duration_stays'] > 14, 15,
                                                   df2['total_duration_stays'])
                                                   )
df2_grpby = df2.groupby(['hotel', 'is_canceled', 'total_duration_stays_group'])['lead_time'].count().reset_index(name='total_bookings')
df2_grpby
```

```python
df2_pivot = df2_grpby.pivot_table(index=['hotel', 'total_duration_stays_group'], columns='is_canceled', values='total_bookings').reset_index()
df2_pivot['ratio_pembatalan'] = round(df2_pivot[1]*100.0/(df2_pivot[0]+df2_pivot[1]), 2)
df2_pivot.head()
```

```python
plt.figure(figsize=(15,10))

sns.regplot(x='total_duration_stays_group', y='ratio_pembatalan', data=df2_pivot[df2_pivot['hotel'] == 'City Hotel'], scatter=False, label='Trend City Hotel')
sns.regplot(x='total_duration_stays_group', y='ratio_pembatalan', data=df2_pivot[df2_pivot['hotel'] == 'Resort Hotel'], scatter=False, label='Trend Resort Hotel')
sns.barplot(data=df2_pivot, x='total_duration_stays_group', y='ratio_pembatalan', hue='hotel',  alpha=0.75)

plt.tick_params(axis='both', labelsize=14)
plt.legend(title='Hotel Type', title_fontsize=14)
plt.xlabel('Durasi Menginap', fontsize=15)
plt.ylabel('Rasio Pembatalan', fontsize=15)
plt.ylim(0, 80)

plt.text(x=-0.3, y=85, s='Lebih banyak pelanggan yang membatalkan penginapan di City hotel', fontsize=20, fontweight='bold')
plt.text(x=-0.3, y=82, s='Semakin lama penginapan, maka pelanggan cenderung membatalkan pesanannya', fontsize=18)

plt.grid()
plt.tight_layout()
plt.show()
```

##Impact Analysis of Lead Time on Hotel Bookings Cancellation Rate

```python
df3 = clean_data.copy()
```

```python
sns.histplot(data=df3, x="lead_time")
```

```python
df3.lead_time.describe()
```

```python
df3.lead_time = np.where(df3.lead_time > 365, 365, df3.lead_time)
```

```python
def convert_lead_time(lead_time):
  value = lead_time
  if lead_time <= 0:
    value = '0-30'
  elif lead_time <= 31:
    value = '31-60'
  elif lead_time <= 61:
    value = '61-90'
  elif lead_time <= 91:
    value = '91-120'
  elif lead_time <= 121:
    value = '121-150'
  elif lead_time <= 151:
    value = '151-180'
  elif lead_time <= 181:
    value = '181-210'
  elif lead_time <= 211:
    value = '211-240'
  elif lead_time <= 241:
    value = '241-270'
  elif lead_time <= 271:
    value = '271-300'
  elif lead_time <= 301:
    value = '301-330'
  elif lead_time <= 365 :
    value = '331-365'
  return value
```

```python
df3['lead_time_group'] = df3['lead_time'].apply(convert_lead_time)
```

```python
df3[['lead_time', 'lead_time_group']]
```

```python
df3_grpby = df3.groupby(['hotel', 'is_canceled', 'lead_time_group'])['hotel'].count().reset_index(name='total_bookings')
df3_pivot = df3_grpby.pivot_table(index=['hotel','lead_time_group'], columns='is_canceled', values='total_bookings').reset_index()
df3_pivot.sort_values(by=['hotel', 'lead_time_group']).head()
```

```python
df3_pivot['ratio_pembatalan'] = round(df3_pivot[1]*100.0/(df3_pivot[0]+df3_pivot[1]), 2)
```

```python
plt.figure(figsize=(15,10))

sns.barplot(data=df3_pivot, x='lead_time_group', y='ratio_pembatalan', hue='hotel')

plt.tick_params(axis='both', labelsize=14)
plt.legend(title='Hotel Type', title_fontsize=14)
plt.xlabel('Lead Time Group (per 30 days)', fontsize=15)
plt.ylabel('Canceled Ratio', fontsize=15)
plt.ylim(0,80)

plt.axvline(0.5, ls='--', color='red')
plt.text(x=-0.4, y=75, s='1 Month', fontsize=15, color='red')

plt.text(x=-0.9, y=90, s='Kedua hotel mempunyai ratio pembatalan penginapan yang rendah ketika lead time 1 bulan', fontsize=20, fontweight='bold')
plt.text(x=-0.9, y=83, s='City hotel mempunyai ratio pembatalan penginapan yang tinggi ketika lead time 1 tahun\nSedangkan resort hotel cenderung lebih rendah', fontsize=18)

plt.grid()
plt.tight_layout()
plt.show()
```
